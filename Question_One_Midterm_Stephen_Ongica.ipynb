```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('telecom_churn.csv')
df.head()
```

```python
# New attribute churn_num
df['churn_num'] = df['churn'].apply(lambda x: 1 if x else 0)
# Displaying the updated DataFrame
df.head()
```

```python
churn_1 = df[df['churn'] == True]
churn_1.head()
```

```python
df.columns
```

```python
# Selecting four numeric attributes
attributes = ['total day minutes', 'total day charge', 'total eve minutes', 'total night minutes']

# Computing the correlation matrix
correlation_matrix = df[attributes].corr()

# Plotting the heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Dataset for customer service calls < 4
filtered_df = df[df['customer service calls'] < 4]

plt.figure(figsize=(10, 6))
plt.plot(
    filtered_df['total day calls'],
         filtered_df['total night minutes'],
         color='blue', marker='o', linestyle='-',
         linewidth=2, markersize=6
)

# Line plot labels
plt.title('Relationship Between Total Day Calls and Total Night Minutes (Customer Service Calls < 4)', fontsize=14)
plt.xlabel('Total Day Calls', fontsize=12)
plt.ylabel('Total Night Minutes', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
